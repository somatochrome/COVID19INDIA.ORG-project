```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import date
from datetime import timedelta
import time
%matplotlib inline
from pathlib import Path 
```

```python
#Find yesterday's date and convert it to string
yesterday=date.today()-timedelta(days=1)
ydate=yesterday.strftime("%d-%m-%Y")

#Specify Data folder
data_folder = Path("C:/Users/joses/Nextcloud/JOSEPH/COVID19INDIAORG ANALYSIS/PYTHON APPROACH/PYTHON CODE/data")

#Compute filename
data_file = ydate + ".csv"

#Compute complete filepath for the day
file_path = data_folder / data_file

#Read file to data frame
df=pd.read_csv(file_path)

#Change format of Date Announce to Date-time format
df['Date Announced']= pd.to_datetime(df['Date Announced'],dayfirst=True)
```

```python
#Daily New Cases - India Calculation
daily_counts=df['Date Announced'].value_counts().sort_index()
daily_counts.rename(columns={'Date Announced':'Daily New Cases'}, inplace=True)
daily_counts.index.name='Date'
```

```python
#Calculate and Print Total Confirmed cases
print("Total Confirmed Cases in India:", daily_counts.sum())
```

```python
#Print Daily New Cases in India
print("Daily New Cases - India",daily_counts)
```

```python
#Calculate and print Cumulateive Cases in India
india_cum=daily_counts.cumsum()
print("Cumulative Cases - India", india_cum)
```

```python
#Specify Export folder 
export_folder =Path("C:/Users/joses/Nextcloud/JOSEPH/COVID19INDIAORG ANALYSIS/PYTHON APPROACH/PYTHON CODE/exports")

#File Name for States Daily and file path
fnsd = ydate +"-State Daily.csv"
fpsd = export_folder / fnsd

#File Name for States Cumulative and filepath 
fnsc = ydate +"-State Cumulative.csv"
fpsc = export_folder / fnsc
```

```python
#Calculate Daily New Cases in States and export to CSV
state_daily=pd.DataFrame(df.groupby(['Date Announced','Detected State']).size())
state_daily_pivot=state_daily.pivot_table(index='Date Announced', columns='Detected State', fill_value=0)
state_daily_pivot.to_csv(fpsd)
```

```python
#Calculate Cumulative Cases in States and export to CSV
state_cum=state_daily_pivot.cumsum()
state_cum.to_csv(fpsc)
```

```python
#Plot cumulative Cases in India
india_cum.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Cumulative Cases India')
plt.show()
#plt.savefig('India Cumulative.png')
```

```python
#Plot Daily New Cases in India
daily_counts.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Daily New Cases - India')
plt.show()
```

```python
#Plot Cumulative Cases in States
state_cum.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Cumulative Cases States')
plt.show()
#plt.savefig('States Cumulative.png')
```
